## bubble sort

```python
def bubble_sort(arr):
    
    swap_occur = True
    j, k = 0, 0
    while swap_occur:
        j += 1
        swap_occur = False
        
        for i in range(len(arr)-1):
            if arr[i] > arr[i+1]:
                k += 1
                arr[i], arr[i+1] = arr[i+1], arr[i]
                swap_occur = True
    return arr, (j+k)
            
            
```

```python
arr = [10, 8, 6, 4, 2, 1]
```

```python
bubble_sort(arr)
```

```python
size = []
iters = []

for i in range(len(arr)):
    for j in range(len(arr)):
        arr[j] = len(arr) - j
    result, count = bubble_sort(arr)
    iters.append(count)
    size.append(len(arr))
    arr += arr
```

```python
plot(size, iters)
```
